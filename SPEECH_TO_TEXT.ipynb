```python
!pip install openai-whisper
!pip install gradio
import whisper
import gradio as gr

# Load the Whisper model (options: "tiny", "base", "small", "medium", "large")
model = whisper.load_model("medium")

def transcribe_audio(file_path, translate=True):
    """
    Converts speech to text using Whisper.
    If translate=True, it translates non-English speech to English.
    """
    try:
        # Transcribe or translate
        task_type = "translate" if translate else "transcribe"
        result = model.transcribe(file_path, task=task_type)

        return result["text"]
    except Exception as e:
        return f"Error: {e}"

# Gradio UI for uploading files
def process_audio(audio_file, translate):
    return transcribe_audio(audio_file, translate)

# Create Gradio interface
iface = gr.Interface(
    fn=process_audio,
    inputs=[
        gr.Audio(type="filepath", label="Upload an Audio File"),
        gr.Checkbox(label="Translate to English?"),
    ],
    outputs="text",
    title="Speech-to-Text Transcription",
    description="Upload an audio file and get the transcription. Enable translation for English output."
)

# Launch the app
iface.launch()
```
